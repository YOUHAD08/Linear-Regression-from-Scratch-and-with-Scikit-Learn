```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, median_absolute_error , r2_score
```

```python
data = pd.read_csv('Adv.csv', index_col=0)
data.head()
```

```python
X,y=data[['TV','Radio','Newspaper']],data['Sales']

learning_rate = 0.00001
number_of_iterations = 1000
W1,W2,W3,bias=1,1,1,0
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=33)
print(y_train)
n=X_train.shape[0]
```

```python
errors=[]
for k in range(number_of_iterations):
    y_hate = W1 * X_train['TV'] + W2 * X_train['Radio'] + W3 * X_train['Newspaper'] + bias  
    error = y_hate - y_train  
    errors.append(np.mean(error ** 2))  
    # 

    dW1 = -(2/n) * np.sum(X_train['TV'] * (y_train - y_hate))
    dW2 = -(2/n) * np.sum(X_train['Radio'] * (y_train - y_hate))
    dW3 = -(2/n) * np.sum(X_train['Newspaper'] * (y_train - y_hate))
    dbias = -(2/n) * np.sum(y_train - y_hate)
     
    W1 =W1 -learning_rate * dW1
    W2 =W2 - learning_rate * dW2
    W3 =W3 - learning_rate * dW3
    bias =bias - learning_rate * dbias
    print(f'itiration : {k}-----W1: {W1} --- W2: {W2} --- W3: {W3} --- bias: {bias}')

print(f'W1: {W1} --- W2: {W2} --- W3: {W3} --- bias: {bias}')
```

```python
y_hate = W1 * X_test['TV'] + W2 * X_test['Radio'] + W3 * X_test['Newspaper'] + bias  
MSE = (1 / n) * np.sum((y_hate - y_test)**2)
MAE = (1 / n) * np.sum(np.abs(y_hate - y_test))
SS_res = np.sum((y_test - y_hate) ** 2)  
SS_tot = np.sum((y_test - np.mean(y_test)) ** 2)
R2 = 1 - (SS_res / SS_tot)
print(f'Mean Square Error: {MSE}')
print(f'Mean Absolute Error : {MAE}')
print(f"R: {R2}")
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
W,bias=model.coef_,model.intercept_
y_hate=model.predict(X_test)
mse=mean_squared_error(y_hate,y_test)
mae=median_absolute_error(y_hate,y_test)
r2=r2_score(y_hate,y_test)
print(f'W1 : {W[0]} --- W2 : {W[1]} --- W3 : {W[2]} --- Bias : {bias}')
print(f'Mean Square Error: {mse}')
print(f'Mean Absolute Error : {mae}')
print(f"R: {r2}")
```
